# Dataset Analysis

```python
import pandas as pd

df = pd.read_csv('E:\\Users\\Prajj\\Documents\\7th Sem\\RM\\Datasets\\Polish Dataset\\Polish Dataset.csv')

result = df.groupby(['Sum_QCHAT', 'group']).size().unstack(fill_value=0)
print(result)

```

```python
import matplotlib.pyplot as plt

# Plot 1: Sum_QCHAT vs Count for group 1
plt.figure(figsize=(8, 4))
result[1].plot(kind='line', marker='o', color='blue')
plt.xlabel('Sum_QCHAT')
plt.ylabel('Count (Group 1)')
plt.title('Sum_QCHAT vs Count for Group 1')
plt.grid(True)
plt.show()

# Plot 2: Sum_QCHAT vs Count for group 7
plt.figure(figsize=(8, 4))
result[7].plot(kind='line', marker='o', color='green')
plt.xlabel('Sum_QCHAT')
plt.ylabel('Count (Group 7)')
plt.title('Sum_QCHAT vs Count for Group 7')
plt.grid(True)
plt.show()
```
